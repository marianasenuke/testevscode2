```python
#Passeio aleatrio (implementao bsica)
import numpy as np
import matplotlib.pyplot as plt

def passeio_aleatorio(N, p):
    """

    """
    rng = np.random.default_rng()

    # Gera N nmeros aleatrios entre 0 e 1
    aleatorios = rng.random(N)

    # Determina a direo de cada passo (1 para direita, -1 para esquerda)
    passos = np.where(aleatorios > p, 1, -1)

    # Calcula a posio cumulativa (soma acumulada dos passos)
    posicao = np.cumsum(passos)

    return posicao

def plotar_passeio(N, p, ax, label):
    """

    """
    posicao = passeio_aleatorio(N, p)
    ax.plot(posicao, label=label, color='pink')
    ax.set_xlabel('Nmero de passos (tempo)')
    ax.set_ylabel('Posio')
    ax.set_title(f'Passeio Aleatrio - p = {p}')
    ax.legend()
    ax.grid(True)

# Parmetros da simulao
N = 1000  # Nmero de passos
p1 = 0.5  # Probabilidade para o primeiro caso
p2 = 0.6  # Probabilidade para o segundo caso

# Cria a figura e os eixos
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Simula e plota os dois casos
plotar_passeio(N, p1, ax1, 'p = 0.5')
plotar_passeio(N, p2, ax2, 'p = 0.6')

plt.tight_layout()
plt.show()
```

```python
#Passeio aleatrio (histograma):

import numpy as np
import matplotlib.pyplot as plt

# Parmetros da simulao
n_passos = 1000  # Nmero total de passos
n_caminhantes = 1000  # Nmero de caminhantes
probabilidade_direita = 0.5  # Probabilidade de ir para a direita

# Simulao do passeio aleatrio
posicoes = np.zeros((n_caminhantes, n_passos))  # Matriz para armazenar as posies

for i in range(1, n_passos):
    # Gera passos aleatrios (-1 ou 1)
    passos = np.random.choice([-1, 1], size=n_caminhantes, p=[1-probabilidade_direita, probabilidade_direita])
    posicoes[:, i] = posicoes[:, i-1] + passos

# Cria histograma para cada passo
n_bins = 50  # Nmero de bins para o histograma
histograma = np.zeros((n_bins, n_passos))  # Matriz para armazenar os histogramas

# Define os limites dos bins
min_pos = posicoes.min()
max_pos = posicoes.max()
bins = np.linspace(min_pos, max_pos, n_bins+1)

for i in range(n_passos):
    hist, _ = np.histogram(posicoes[:, i], bins=bins)
    histograma[:, i] = hist

# Plot do histograma no ltimo passo
plt.figure(figsize=(10, 5))
plt.hist(posicoes[:, -1], bins=bins, color='blue', alpha=0.7)
plt.title(f'Histograma da Posio dos Caminhantes no Passo {n_passos}')
plt.xlabel('Posio')
plt.ylabel('Nmero de Caminhantes')
plt.grid(True)
plt.show()

# Plot do mapa de cores da evoluo do histograma
plt.figure(figsize=(12, 6))
plt.imshow(histograma, aspect='auto', cmap='hot',
           extent=[0, n_passos, min_pos, max_pos], origin='lower')
plt.colorbar(label='Nmero de Caminhantes')
plt.title('Evoluo do Histograma da Posio dos Caminhantes')
plt.xlabel('Passo')
plt.ylabel('Posio')
plt.show()
```

```python

#Artigo:
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

D = 0.1
L = 10            # Tamanho
total_time = 2.0  # Tempo total de simulao
dt = 0.01

Nx = 100
Nt = int(total_time / dt)
N_particles = 1000


x = np.linspace(0, L, Nx)
dx = x[1] - x[0]

# Condio inicial (Gaussiana centrada)
n = np.exp(-(x - L/2)**2 / 0.5)
n = n / (np.sum(n)*dx)  # Normalizao

# Evoluo temporal
for _ in range(Nt):
    n[1:-1] += D * dt / dx**2 * (n[2:] - 2 * n[1:-1] + n[:-2])


positions = np.zeros(N_particles) + L/2

for _ in range(Nt):
    positions += np.random.normal(0, np.sqrt(2 * D * dt), N_particles)


plt.figure(figsize=(12, 6))


plt.plot(x, n, 'r-', linewidth=2, label='Soluo da Equao de Difuso')


plt.hist(positions, bins=30, density=True, alpha=0.5,
         color='b', label='Simulao Estocstica')


if D > 0:
    sigma = np.sqrt(2 * D * total_time)
    analytical = norm.pdf(x, L/2, sigma)
    plt.plot(x, analytical, 'g--', label='Soluo Analtica')

plt.xlabel('Posio (x)')
plt.ylabel('Densidade de Probabilidade')
plt.title(f'Difuso de Partculas (D = {D}, t = {total_time})')
plt.legend()
plt.grid(True)
plt.show()
```
