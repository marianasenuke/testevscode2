```python
#Passeio aleatório (implementação básica)
import numpy as np
import matplotlib.pyplot as plt

def passeio_aleatorio(N, p):
    """

    """
    rng = np.random.default_rng()

    # Gera N números aleatórios entre 0 e 1
    aleatorios = rng.random(N)

    # Determina a direção de cada passo (1 para direita, -1 para esquerda)
    passos = np.where(aleatorios > p, 1, -1)

    # Calcula a posição cumulativa (soma acumulada dos passos)
    posicao = np.cumsum(passos)

    return posicao

def plotar_passeio(N, p, ax, label):
    """

    """
    posicao = passeio_aleatorio(N, p)
    ax.plot(posicao, label=label, color='pink')
    ax.set_xlabel('Número de passos (tempo)')
    ax.set_ylabel('Posição')
    ax.set_title(f'Passeio Aleatório - p = {p}')
    ax.legend()
    ax.grid(True)

# Parâmetros da simulação
N = 1000  # Número de passos
p1 = 0.5  # Probabilidade para o primeiro caso
p2 = 0.6  # Probabilidade para o segundo caso

# Cria a figura e os eixos
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Simula e plota os dois casos
plotar_passeio(N, p1, ax1, 'p = 0.5')
plotar_passeio(N, p2, ax2, 'p = 0.6')

plt.tight_layout()
plt.show()
```

```python
#Passeio aleatório (histograma):

import numpy as np
import matplotlib.pyplot as plt

# Parâmetros da simulação
n_passos = 1000  # Número total de passos
n_caminhantes = 1000  # Número de caminhantes
probabilidade_direita = 0.5  # Probabilidade de ir para a direita

# Simulação do passeio aleatório
posicoes = np.zeros((n_caminhantes, n_passos))  # Matriz para armazenar as posições

for i in range(1, n_passos):
    # Gera passos aleatórios (-1 ou 1)
    passos = np.random.choice([-1, 1], size=n_caminhantes, p=[1-probabilidade_direita, probabilidade_direita])
    posicoes[:, i] = posicoes[:, i-1] + passos

# Cria histograma para cada passo
n_bins = 50  # Número de bins para o histograma
histograma = np.zeros((n_bins, n_passos))  # Matriz para armazenar os histogramas

# Define os limites dos bins
min_pos = posicoes.min()
max_pos = posicoes.max()
bins = np.linspace(min_pos, max_pos, n_bins+1)

for i in range(n_passos):
    hist, _ = np.histogram(posicoes[:, i], bins=bins)
    histograma[:, i] = hist

# Plot do histograma no último passo
plt.figure(figsize=(10, 5))
plt.hist(posicoes[:, -1], bins=bins, color='blue', alpha=0.7)
plt.title(f'Histograma da Posição dos Caminhantes no Passo {n_passos}')
plt.xlabel('Posição')
plt.ylabel('Número de Caminhantes')
plt.grid(True)
plt.show()

# Plot do mapa de cores da evolução do histograma
plt.figure(figsize=(12, 6))
plt.imshow(histograma, aspect='auto', cmap='hot',
           extent=[0, n_passos, min_pos, max_pos], origin='lower')
plt.colorbar(label='Número de Caminhantes')
plt.title('Evolução do Histograma da Posição dos Caminhantes')
plt.xlabel('Passo')
plt.ylabel('Posição')
plt.show()
```

```python

#Artigo:
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

D = 0.1
L = 10            # Tamanho
total_time = 2.0  # Tempo total de simulação
dt = 0.01

Nx = 100
Nt = int(total_time / dt)
N_particles = 1000


x = np.linspace(0, L, Nx)
dx = x[1] - x[0]

# Condição inicial (Gaussiana centrada)
n = np.exp(-(x - L/2)**2 / 0.5)
n = n / (np.sum(n)*dx)  # Normalização

# Evolução temporal
for _ in range(Nt):
    n[1:-1] += D * dt / dx**2 * (n[2:] - 2 * n[1:-1] + n[:-2])


positions = np.zeros(N_particles) + L/2

for _ in range(Nt):
    positions += np.random.normal(0, np.sqrt(2 * D * dt), N_particles)


plt.figure(figsize=(12, 6))


plt.plot(x, n, 'r-', linewidth=2, label='Solução da Equação de Difusão')


plt.hist(positions, bins=30, density=True, alpha=0.5,
         color='b', label='Simulação Estocástica')


if D > 0:
    sigma = np.sqrt(2 * D * total_time)
    analytical = norm.pdf(x, L/2, sigma)
    plt.plot(x, analytical, 'g--', label='Solução Analítica')

plt.xlabel('Posição (x)')
plt.ylabel('Densidade de Probabilidade')
plt.title(f'Difusão de Partículas (D = {D}, t = {total_time})')
plt.legend()
plt.grid(True)
plt.show()
```
