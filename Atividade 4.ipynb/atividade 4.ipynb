```python
#Passeio aleatório (implementação básica)
import numpy as np
import matplotlib.pyplot as plt

def passeio_aleatorio(N, p):
    """

    """
    rng = np.random.default_rng()

    # Gera N números aleatórios entre 0 e 1
    aleatorios = rng.random(N)

    # Determina a direção de cada passo (1 para direita, -1 para esquerda)
    passos = np.where(aleatorios > p, 1, -1)

    # Calcula a posição cumulativa (soma acumulada dos passos)
    posicao = np.cumsum(passos)

    return posicao

def plotar_passeio(N, p, ax, label):
    """

    """
    posicao = passeio_aleatorio(N, p)
    ax.plot(posicao, label=label, color='pink')
    ax.set_xlabel('Número de passos (tempo)')
    ax.set_ylabel('Posição')
    ax.set_title(f'Passeio Aleatório - p = {p}')
    ax.legend()
    ax.grid(True)

# Parâmetros da simulação
N = 1000  # Número de passos
p1 = 0.5  # Probabilidade para o primeiro caso
p2 = 0.6  # Probabilidade para o segundo caso

# Cria a figura e os eixos
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Simula e plota os dois casos
plotar_passeio(N, p1, ax1, 'p = 0.5')
plotar_passeio(N, p2, ax2, 'p = 0.6')

plt.tight_layout()
plt.show()
```

```python
#Passeio aleatório (repetição):
import numpy as np
import matplotlib.pyplot as plt

def passeio_aleatorio(N, p):
    """
    Simula um passeio aleatório unidimensional.

    Parâmetros:
    N (int): Número total de passos
    p (float): Probabilidade de dar um passo para a direita

    Retorna:
    numpy.ndarray: Array com a posição do caminhante em cada passo
    """
    rng = np.random.default_rng()
    aleatorios = rng.random(N)
    passos = np.where(aleatorios < p, 1, -1)
    return np.cumsum(passos)

def plotar_multiplos_passeios(N, p, num_simulacoes, ax, title):
    """
    Simula e plota múltiplos passeios aleatórios.
    """
    for _ in range(num_simulacoes):
        posicao = passeio_aleatorio(N, p)
        ax.plot(posicao, alpha=0.5, linewidth=0.8)

    ax.set_xlabel('Número de passos (tempo)')
    ax.set_ylabel('Posição')
    ax.set_title(title)
    ax.grid(True)

# Parâmetros da simulação
N = 1000  # Número de passos
num_simulacoes = 100  # Número de simulações para cada p
p1 = 0.5  # Probabilidade para o primeiro caso
p2 = 0.6  # Probabilidade para o segundo caso

# Cria a figura e os eixos
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Simula e plota os dois casos com múltiplas simulações
plotar_multiplos_passeios(N, p1, num_simulacoes, ax1, f'100 Passeios Aleatórios com p = {p1}')
plotar_multiplos_passeios(N, p2, num_simulacoes, ax2, f'100 Passeios Aleatórios com p = {p2}')

plt.tight_layout()
plt.show()
```

```python
#Passeio aleatório (médias):
import numpy as np
import matplotlib.pyplot as plt

def passeio_aleatorio(N, p):
    """

    """
    rng = np.random.default_rng()  # Cria um gerador de números aleatórios

    # Gera N números aleatórios entre 0 e 1
    aleatorios = rng.random(N)

    # Determina a direção de cada passo (1 para direita, -1 para esquerda)
    passos = np.where(aleatorios < p, 1, -1)

    # Calcula a posição cumulativa (soma acumulada dos passos)
    posicao = np.cumsum(passos)

    return posicao

# Parâmetros da simulação
N = 1000  # Número de passos
p1 = 0.5  # Probabilidade para o primeiro caso
p2 = 0.6  # Probabilidade para o segundo caso

# Simula os passeios
passeio1 = passeio_aleatorio(N, p1)
passeio2 = passeio_aleatorio(N, p2)

# Configuração do gráfico
plt.figure(figsize=(10, 6))

# Plota os dois passeios
plt.plot(passeio1, label=f'p = {p1}', alpha=0.8)
plt.plot(passeio2, label=f'p = {p2}', alpha=0.8)

# Adiciona elementos ao gráfico
plt.xlabel('Número de passos (tempo)')
plt.ylabel('Posição')
plt.title('Passeio Aleatório Unidimensional')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Mostra o gráfico
plt.show()
```

```python
# Passeio aleatório (histograma 1):
import numpy as np
import matplotlib.pyplot as plt

def passeio_aleatorio(N, p):
    """Simula um passeio aleatório unidimensional."""
    rng = np.random.default_rng()
    aleatorios = rng.random(N)
    passos = np.where(aleatorios < p, 1, -1)
    return np.cumsum(passos)

def gerar_histogramas(N, p, num_simulacoes=1000):
    """Gera histogramas para posições inicial, intermediária e final."""
    pos_inicio = []
    pos_meio = []
    pos_final = []

    for _ in range(num_simulacoes):
        passeio = passeio_aleatorio(N, p)
        pos_inicio.append(passeio[0])  # Posição inicial (sempre 0)
        pos_meio.append(passeio[N//2])  # Posição no meio
        pos_final.append(passeio[-1])   # Posição final

    return pos_inicio, pos_meio, pos_final

# Parâmetros da simulação
N = 1000
p1 = 0.5
p2 = 0.6
num_simulacoes = 1000

# Gera dados para p = 0.5
inicio1, meio1, final1 = gerar_histogramas(N, p1, num_simulacoes)

# Gera dados para p = 0.6
inicio2, meio2, final2 = gerar_histogramas(N, p2, num_simulacoes)

# Configuração dos gráficos
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot para p = 0.5
ax1.hist([inicio1, meio1, final1], bins=30, label=['Início', 'Meio', 'Final'], alpha=0.7)
ax1.set_title(f'Distribuição das Posições (p = {p1})')
ax1.set_xlabel('Posição')
ax1.set_ylabel('Frequência')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot para p = 0.6
ax2.hist([inicio2, meio2, final2], bins=30, label=['Início', 'Meio', 'Final'], alpha=0.7)
ax2.set_title(f'Distribuição das Posições (p = {p2})')
ax2.set_xlabel('Posição')
ax2.set_ylabel('Frequência')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
