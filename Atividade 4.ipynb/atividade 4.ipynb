```python
#Passeio aleatrio (implementao bsica)
import numpy as np
import matplotlib.pyplot as plt

def passeio_aleatorio(N, p):
    """

    """
    rng = np.random.default_rng()

    # Gera N nmeros aleatrios entre 0 e 1
    aleatorios = rng.random(N)

    # Determina a direo de cada passo (1 para direita, -1 para esquerda)
    passos = np.where(aleatorios > p, 1, -1)

    # Calcula a posio cumulativa (soma acumulada dos passos)
    posicao = np.cumsum(passos)

    return posicao

def plotar_passeio(N, p, ax, label):
    """

    """
    posicao = passeio_aleatorio(N, p)
    ax.plot(posicao, label=label, color='pink')
    ax.set_xlabel('Nmero de passos (tempo)')
    ax.set_ylabel('Posio')
    ax.set_title(f'Passeio Aleatrio - p = {p}')
    ax.legend()
    ax.grid(True)

# Parmetros da simulao
N = 1000  # Nmero de passos
p1 = 0.5  # Probabilidade para o primeiro caso
p2 = 0.6  # Probabilidade para o segundo caso

# Cria a figura e os eixos
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Simula e plota os dois casos
plotar_passeio(N, p1, ax1, 'p = 0.5')
plotar_passeio(N, p2, ax2, 'p = 0.6')

plt.tight_layout()
plt.show()
```

```python
#Passeio aleatrio (repetio):
import numpy as np
import matplotlib.pyplot as plt

def passeio_aleatorio(N, p):
    """
    Simula um passeio aleatrio unidimensional.

    Parmetros:
    N (int): Nmero total de passos
    p (float): Probabilidade de dar um passo para a direita

    Retorna:
    numpy.ndarray: Array com a posio do caminhante em cada passo
    """
    rng = np.random.default_rng()
    aleatorios = rng.random(N)
    passos = np.where(aleatorios < p, 1, -1)
    return np.cumsum(passos)

def plotar_multiplos_passeios(N, p, num_simulacoes, ax, title):
    """
    Simula e plota mltiplos passeios aleatrios.
    """
    for _ in range(num_simulacoes):
        posicao = passeio_aleatorio(N, p)
        ax.plot(posicao, alpha=0.5, linewidth=0.8)

    ax.set_xlabel('Nmero de passos (tempo)')
    ax.set_ylabel('Posio')
    ax.set_title(title)
    ax.grid(True)

# Parmetros da simulao
N = 1000  # Nmero de passos
num_simulacoes = 100  # Nmero de simulaes para cada p
p1 = 0.5  # Probabilidade para o primeiro caso
p2 = 0.6  # Probabilidade para o segundo caso

# Cria a figura e os eixos
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Simula e plota os dois casos com mltiplas simulaes
plotar_multiplos_passeios(N, p1, num_simulacoes, ax1, f'100 Passeios Aleatrios com p = {p1}')
plotar_multiplos_passeios(N, p2, num_simulacoes, ax2, f'100 Passeios Aleatrios com p = {p2}')

plt.tight_layout()
plt.show()
```

```python
#Passeio aleatrio (mdias):
import numpy as np
import matplotlib.pyplot as plt

def passeio_aleatorio(N, p):
    """

    """
    rng = np.random.default_rng()  # Cria um gerador de nmeros aleatrios

    # Gera N nmeros aleatrios entre 0 e 1
    aleatorios = rng.random(N)

    # Determina a direo de cada passo (1 para direita, -1 para esquerda)
    passos = np.where(aleatorios < p, 1, -1)

    # Calcula a posio cumulativa (soma acumulada dos passos)
    posicao = np.cumsum(passos)

    return posicao

# Parmetros da simulao
N = 1000  # Nmero de passos
p1 = 0.5  # Probabilidade para o primeiro caso
p2 = 0.6  # Probabilidade para o segundo caso

# Simula os passeios
passeio1 = passeio_aleatorio(N, p1)
passeio2 = passeio_aleatorio(N, p2)

# Configurao do grfico
plt.figure(figsize=(10, 6))

# Plota os dois passeios
plt.plot(passeio1, label=f'p = {p1}', alpha=0.8)
plt.plot(passeio2, label=f'p = {p2}', alpha=0.8)

# Adiciona elementos ao grfico
plt.xlabel('Nmero de passos (tempo)')
plt.ylabel('Posio')
plt.title('Passeio Aleatrio Unidimensional')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Mostra o grfico
plt.show()
```

```python
# Passeio aleatrio (histograma 1):
import numpy as np
import matplotlib.pyplot as plt

def passeio_aleatorio(N, p):
    """Simula um passeio aleatrio unidimensional."""
    rng = np.random.default_rng()
    aleatorios = rng.random(N)
    passos = np.where(aleatorios < p, 1, -1)
    return np.cumsum(passos)

def gerar_histogramas(N, p, num_simulacoes=1000):
    """Gera histogramas para posies inicial, intermediria e final."""
    pos_inicio = []
    pos_meio = []
    pos_final = []

    for _ in range(num_simulacoes):
        passeio = passeio_aleatorio(N, p)
        pos_inicio.append(passeio[0])  # Posio inicial (sempre 0)
        pos_meio.append(passeio[N//2])  # Posio no meio
        pos_final.append(passeio[-1])   # Posio final

    return pos_inicio, pos_meio, pos_final

# Parmetros da simulao
N = 1000
p1 = 0.5
p2 = 0.6
num_simulacoes = 1000

# Gera dados para p = 0.5
inicio1, meio1, final1 = gerar_histogramas(N, p1, num_simulacoes)

# Gera dados para p = 0.6
inicio2, meio2, final2 = gerar_histogramas(N, p2, num_simulacoes)

# Configurao dos grficos
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot para p = 0.5
ax1.hist([inicio1, meio1, final1], bins=30, label=['Incio', 'Meio', 'Final'], alpha=0.7)
ax1.set_title(f'Distribuio das Posies (p = {p1})')
ax1.set_xlabel('Posio')
ax1.set_ylabel('Frequncia')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot para p = 0.6
ax2.hist([inicio2, meio2, final2], bins=30, label=['Incio', 'Meio', 'Final'], alpha=0.7)
ax2.set_title(f'Distribuio das Posies (p = {p2})')
ax2.set_xlabel('Posio')
ax2.set_ylabel('Frequncia')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
